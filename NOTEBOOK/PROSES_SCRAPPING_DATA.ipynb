# ***MENGGABUNGKAN SELURUH FILE CSV KEDALAM 1 FILE***

```python
import pandas as pd
import glob
import os

# =========================
# 1. PATH FOLDER
# =========================
folder_path = "/content/drive/MyDrive/SKRIPSI/DATA_SET/"

all_files = glob.glob(os.path.join(folder_path, "*.csv"))
print("Jumlah file ditemukan:", len(all_files))

# =========================
# 2. GABUNGKAN SEMUA FILE
# =========================
df_list = []

for file in all_files:
    print("Membaca:", file)

    df = pd.read_csv(
        file,
        sep=",",
        quotechar='"',
        engine="python"
    )

    df_list.append(df)

combined_df = pd.concat(df_list, ignore_index=True)

print("Total baris setelah digabung:", combined_df.shape)

# =========================
# 3. UBAH KE DATETIME
# =========================
combined_df['created_at'] = pd.to_datetime(
    combined_df['created_at'],
    format='%a %b %d %H:%M:%S %z %Y',
    errors='coerce'
)

# Hapus jika ada tanggal gagal terbaca
combined_df = combined_df.dropna(subset=['created_at'])

# =========================
# 4. FILTER RENTANG TANGGAL
# =========================
filtered_df = combined_df[
    (combined_df['created_at'] >= '2025-11-01') &
    (combined_df['created_at'] <= '2026-01-16')
]

# =========================
# 5. URUTKAN TANGGAL
# =========================
sorted_df = filtered_df.sort_values(by='created_at')

# =========================
# 6. SIMPAN KE DRIVE
# =========================
output_path = "/content/drive/MyDrive/SKRIPSI/DATA_SET/DATASET_MBG_FINAL.csv"
sorted_df.to_csv(output_path, index=False)

print("SELESAI!")
print("Total data final:", sorted_df.shape)
```
